## SÉRIES ESTATÍSTICAS
Na estatística é comum armazenar os dados na ordem que eles foram coletados, o que causa uma desorganização e dificuldade na compreensão.

É qualquer método de visualização *(tabelas e gráficos)* que apresenta a distribuição de um conjunto de dados estatísticos em função da época, do local e da espécie.

As séries estatísticas possuem as seguintes classificações:
* __Série Cronológica:__ É a série cujos dados estão dispostos em correspondência com o tempo, ou seja, varia o tempo e permanece constante o fato e o local
* __Série Territorial:__ É a série cujos dados estão dispostos em correspondência com o local, ou seja, varia o local e permanece o tempo e o fato.
* __Série Qualitativa:__  É a série cujos dados estão dispostos em correspondência com a espécie ou qualidade, ou seja, varia o fato e permanece constante a época e o local.
* __Série Composta:__ A combinação entre duas ou mais séries constituem novas séries denominadas compostas.

## ELEMENTOS DE UMA TABELA

Na estatística é importante sintetizar os valores que uma ou mais variáveis podem assumir. Esses valores são apresentados em forma de tabelas e gráficos, nas quais nos mostram as informações a respeito da variável de uma forma rápida e de fácil compreensão, permitindo fazermos escolar mais coerentes.A tabela possui os seguintes elementos:
* Título: Deve responder as questões: O que?, Onde? e Quando
* Cabeçalho: Parte da tabela na qual é designada a natureza do conteúdo de cada coluna.
* Corpo: Parte da tabela composta por linhas, colunas e células.
* Rodapé: Espaço da tabela para colocar notas de natureza informativas (fonte, notas e chamadas).

## GRÁFICOS

Um gráfico corresponde à representação dos dados sob diferentes formas gráficas, a fim de permitir uma visão rápida e global do fato estudado.

De uma maneira geral, os gráficos devem ser elaborados de maneira simples e clara, de tal forma que o observador entenda claramente aquilo que o gráfico busca evidenciar, sem necessidade de ficar procurando adivinhar o que ele representa.
